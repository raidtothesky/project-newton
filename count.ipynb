```python
from ultralytics import YOLO
import matplotlib.pyplot as plt

def count_apples(model='high-bias.pt',plot=False,imagepath="https://storage.googleapis.com/648010c1-f244-4641-98f2-73ff6c1b4e99/ai_assignment_20230726_count.jpeg"):
  yolo_model = YOLO("high-bias.pt")
  results = yolo_model(imagepath)
  if plot:
    plt.imshow(results[0].plot()[:,:,::-1]) # swap BGR to RGB
    plt.show()
  return results[0].__len__()
```

```python
print("The number of apples is {}".format(count_apples(plot=True))) # you can try changing the model and/or image
```
